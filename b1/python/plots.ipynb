```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 100, 1000)
offset = np.array(1000 * [9])
data = np.sin(time)
```

```python
plt.rcParams.update({'font.size': 16})  # Set global font size
plt.plot(time, data, label=r"$v_{\mathrm{gs}}$", lw=3)
plt.plot(time, data + offset, label=r"$v_{\mathrm{GS}}$", lw=3)
plt.plot(time, offset, label=r"$V_{\mathrm{GS}}$", lw=3)
plt.legend()
plt.xlim(0, 100)
plt.ylim(-1, 11)
plt.gca().spines['top'].set_visible(False)   # Remove the top border
plt.gca().spines['right'].set_visible(False) # Remove the right border
plt.gca().spines['bottom'].set_position('zero') # Move bottom border to y=0
plt.gca().spines['left'].set_linewidth(2)    # Make left axis thicker
plt.gca().spines['bottom'].set_linewidth(2)  # Make bottom axis thicker
plt.tick_params(axis='both', which='major', width=2, length=6)  # Make ticks thicker
plt.xticks([])
plt.xlabel("time", labelpad=30)
plt.ylabel("voltage")  # Add a top margin to the ylabel
plt.savefig("../plots/notational_convention.pdf")
```
